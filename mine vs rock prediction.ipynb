```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Loading the Data**

```python
file_path = "sonar data.csv"  # relative path
sonar_data = pd.read_csv(file_path ,header=None)
```

**M--> Mine**
**R--> Rock**

```python
sonar_data.groupby(60).mean()
```

```python
#seperating data and labels
X = sonar_data.drop(columns=60 , axis = 1)
Y = sonar_data[60]
```

```python
X.shape
```

```python
Y.shape
```

```python
print (X)
print(Y)
```

**Splitting the data into Training and Testing**

```python
X_train ,X_test ,Y_train ,Y_test = train_test_split (X,Y,test_size=0.1 ,stratify=Y ,random_state =1)
```

```python
print (X.shape ,X_train.shape ,X_test.shape)
```

**Mosel Training -- Logistic regression Model**

```python
model = LogisticRegression()
```

```python
#training logistic regression model using training data
model.fit(X_train , Y_train)
```

**Model Evaluation**

```python
#accuracy of the training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print ("Accuracy on training data",training_data_accuracy)
```

```python
#accuracy of the testing data 
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print ("Accuracy on testing data",test_data_accuracy)
```
